```python

import numpy as np
import matplotlib.pyplot as plt
import math
from statistics import NormalDist

np.random.seed(42)
```

# 1️⃣ Distribuciones Discretas

### Definición general
Una **distribución discreta** describe variables aleatorias que solo pueden tomar **valores contables**, por ejemplo:
0, 1, 2, 3, ...

No existen valores intermedios.

---


## 1.1 Distribución Binomial

### 📌 Definición teórica
La distribución binomial modela el **número de éxitos** en un número fijo de intentos independientes, donde:
- Cada intento tiene solo dos resultados posibles (éxito / fracaso)
- La probabilidad de éxito es constante

Parámetros:
- **n**: número de intentos  
- **p**: probabilidad de éxito  
- **k**: número de éxitos observados

---
### 📍 Ejemplos reales de uso
1. Número de clientes que **entran en default** en un portafolio de 100 clientes  
2. Número de transacciones **fraudulentas** en un lote fijo de transacciones  
3. Número de estudiantes que **aprueban** un examen (sí / no)

---


```python

def binomial_pmf(k, n, p):
    return math.comb(n, k) * (p**k) * ((1-p)**(n-k))

n = 10
p = 0.2

ks = list(range(0, n+1))
probs = [binomial_pmf(k, n, p) for k in ks]

plt.figure()
plt.bar(ks, probs)
plt.title("Distribución Binomial (n=10, p=0.2)")
plt.xlabel("Número de éxitos")
plt.ylabel("Probabilidad")
plt.show()

sum(probs)
```

```python

# Ejemplo: probabilidad de exactamente 2 éxitos
binomial_pmf(2, n, p)
```

## 1.2 Distribución Poisson

### 📌 Definición teórica
La distribución de Poisson modela el **número de veces que ocurre un evento** en un intervalo fijo de tiempo o espacio, cuando:
- Los eventos ocurren de forma aleatoria
- El promedio de ocurrencias es constante

Parámetro:
- **λ (lambda)**: número promedio de eventos

---
### 📍 Ejemplos reales de uso
1. Número de **pagos tarde** de un cliente en un año  
2. Número de **llamadas** que recibe un call center por hora  
3. Número de **fallas** en un sistema durante un día

---


```python

def poisson_pmf(k, lam):
    return math.exp(-lam) * (lam**k) / math.factorial(k)

lam = 1.5

ks = list(range(0, 10))
probs = [poisson_pmf(k, lam) for k in ks]

plt.figure()
plt.bar(ks, probs)
plt.title("Distribución Poisson (λ = 1.5)")
plt.xlabel("Número de eventos")
plt.ylabel("Probabilidad")
plt.show()

sum(probs)
```

```python

# Ejemplo: probabilidad de 0 eventos
poisson_pmf(0, lam)
```

# 2️⃣ Distribuciones Continuas

### Definición general
Una **distribución continua** describe variables aleatorias que pueden tomar **infinitos valores posibles** dentro de un intervalo.

Ejemplos:
- Ingreso
- Tiempo
- Altura
- Temperatura

---


## 2.1 Distribución Normal

### 📌 Definición teórica
La distribución normal describe variables que:
- Se concentran alrededor de un promedio
- Son simétricas
- Tienen forma de campana

Parámetros:
- **μ (media)**  
- **σ (desviación estándar)**

---
### 📍 Ejemplos reales de uso
1. Errores de medición  
2. Variables transformadas (log-ingresos, residuos de modelos)  
3. Resultados agregados de procesos naturales

---


```python

mu = 50
sigma = 10

x = np.random.normal(mu, sigma, 5000)

plt.figure()
plt.hist(x, bins=40)
plt.title("Distribución Normal (μ=50, σ=10)")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()

x.mean(), x.std()
```

### Regla empírica
- 68% de los valores están dentro de ±1σ  
- 95% dentro de ±2σ  
- 99.7% dentro de ±3σ


## 2.2 Normal Estándar y Z-score

### 📌 Definición teórica
El **Z-score** mide cuántas desviaciones estándar se encuentra un valor respecto al promedio.

Fórmula:
Z = (x − μ) / σ

---
### 📍 Ejemplos reales de uso
1. Detección de **outliers**  
2. Comparar valores en diferentes escalas  
3. Cálculo de probabilidades en la normal

---


```python

x_val = 65
z = (x_val - mu) / sigma
z
```

```python

# Probabilidad de observar un valor mayor a x_val
1 - NormalDist().cdf(z)
```

# 3️⃣ Resumen comparativo

| Distribución | Tipo | ¿Qué modela? | Ejemplos |
|-------------|------|--------------|----------|
| Binomial | Discreta | Éxitos en n intentos | Default, fraude |
| Poisson | Discreta | Conteo de eventos | Pagos tarde |
| Normal | Continua | Valores alrededor del promedio | Errores |
| Z-score | Continua | Distancia al promedio | Outliers |
