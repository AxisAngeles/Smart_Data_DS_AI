```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# URL base
url = "https://listado.mercadolibre.com.pe/laptop#D[A:laptop]"
```

```python
# Hacemos la petición
response = requests.get(url)
if response.status_code != 200:
    print("Error al acceder a la página")
    exit()
```

```python
# Parseamos el contenido HTML
soup = BeautifulSoup(response.text, "html.parser")
```

```python
# Buscamos los contenedores de productos
productos = soup.find_all("li", {"class": "ui-search-layout__item"})
```

```python
# Lista para guardar resultados
data = []
for producto in productos:
    # Nombre
    nombre = producto.find("h2", {"class": "ui-search-item__title"})
    nombre = nombre.text if nombre else "No encontrado"

    # Precio
    precio = producto.find("span", {"class": "andes-money-amount__fraction"})
    precio = precio.text if precio else "No disponible"

    # Link
    link = producto.find("a", {"class": "ui-search-link"})
    link = link["href"] if link else "No disponible"

    data.append([nombre, precio, link])
```

```python
# Guardamos en un DataFrame
df = pd.DataFrame(data, columns=["Producto", "Precio", "Link"])
```

```python
# Mostramos resultados
print(df.head())
```

```python
# Exportamos a CSV
df.to_csv("laptops_mercadolibre.csv", index=False, encoding="utf-8-sig")
print("Archivo CSV generado con éxito 🚀")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# URL del portal
url = "https://elcomercio.pe/ultimas-noticias/"
```

```python
# Hacemos la petición HTTP
response = requests.get(url)
if response.status_code != 200:
    print("Error al acceder a la página")
    exit()
```

```python
# Parseamos con BeautifulSoup
soup = BeautifulSoup(response.text, "html.parser")
```

```python
# Contenedor de las noticias
noticias = soup.find_all("div", class_="story-item")
```

```python
# Lista para guardar resultados
data = []

for n in noticias:
    # Título
    titulo = n.find("a", class_="story-item__title")
    titulo = titulo.text.strip() if titulo else "No encontrado"

    # Link
    link = n.find("a", class_="story-item__title")
    link = "https://elcomercio.pe" + link["href"] if link else "No disponible"

    # Resumen (puede no existir)
    resumen = n.find("p", class_="story-item__summary")
    resumen = resumen.text.strip() if resumen else "No disponible"

    data.append([titulo, link, resumen])
```

```python
# Guardamos en un DataFrame
df = pd.DataFrame(data, columns=["Título", "Link", "Resumen"])
```

```python
# Mostramos resultados
print(df.head())
```

```python
# Exportamos a CSV
df.to_csv("noticias_elcomercio.csv", index=False, encoding="utf-8-sig")
print("✅ Archivo CSV generado con éxito")
```
