```python
# 1) Librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, root_mean_squared_error, r2_score


from sklearn.datasets import load_breast_cancer, load_diabetes, fetch_california_housing
# Load and return the breast cancer Wisconsin dataset (classification).
# Load and return the diabetes dataset (regression).
# Load the California housing dataset (regression).
```

```python
# 1) Crear DataFrames
data = load_diabetes()
print(f'Keys: {data.keys()}')

# 1.1) DIABETES
df_diabetes = pd.DataFrame(columns=data.feature_names, data=data.data)
df1 = pd.DataFrame(columns=["target"], data=data.target)
df_diabetes = pd.concat([df_diabetes,df1], axis=1)
print(f'Diabetes shape: {df_diabetes.shape}')

# 1.2) BREAST CANCER
data = load_breast_cancer()
df_breastc = pd.DataFrame(columns=data.feature_names, data=data.data)
df1 = pd.DataFrame(columns=["target"], data=data.target)
df_breastc = pd.concat([df_breastc,df1], axis=1)
print(f'Breast Cancer shape: {df_breastc.shape}')
```
